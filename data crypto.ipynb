```python

!pip install yfinance

import yfinance as yf
import pandas as pd

# Select cryptocurrency
coin = "BTC-USD"  # Change if needed

# Download full data
df = yf.download(
    coin,
    start="2025-01-01",
    end="2025-12-01",   # 30 Nov
    progress=False
)

# Show all rows (Jupyter normally hides rows)
pd.set_option('display.max_rows', None)

df

df.to_csv("BTC_1Jan2025_to_30Nov2025.csv")

print("CSV file created successfully!")
```
